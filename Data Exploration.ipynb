```python
import pandas as pd
import numpy as np
```

```python
train_data_raw = pd.read_csv('data/pml_train.csv')
test_data_raw = pd.read_csv('data/pml_test_features.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(train_data_raw.shape)
print(train_data_raw.columns)
train_data_raw.head()
```

```python
fig = plt.figure(figsize=(100, 100))

for i in range(10):
    for j in range(10):
        plt.subplot(10,10,i*10+j+1)
        index = i*10+j+1
        var = "cat" + str(index)
        train_data_raw[var].value_counts().plot(kind="bar")
```

```python
fig = plt.figure(figsize=(50, 50))
for i in range(4):
    for j in range(4):
        plt.subplot(4,4,i*4+j+1)
        index = 100+i*4+j+1
        var = "cat" + str(index)
        train_data_raw[var].value_counts().plot(kind="bar")
```
